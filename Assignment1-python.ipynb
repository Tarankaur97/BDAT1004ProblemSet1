6) You can turn a word into pig-Latin using the following two rules (simplified):• If the word starts with a consonant, move that letter to the end and append 'ay'. For example, 'happy' becomes 'appyhay' and 'pencil' becomes 'encilpay'.• If the word starts with a vowel, simply append 'way' to the end of the word. For example, 'enter' becomes 'enterway' and 'other' becomes 'otherway' . For our purposes, there are 5 vowels: a, e, i, o, u (so we count y as a consonant).Write a function pig() that takes a word (i.e., a string) as input and returns its pig- Latin form. Your function should still work if the input word contains upper case characters. Your output should always be lower case however.

```python
vowel_array = ['a','e','i','o','u']
def pig(word):
    
    #if word starts with vowel or not
    if word[0].lower() in vowel_array:
        #print the output
        print(word.lower() + 'way') 
    else:
        first_letter = word[0]
        print(word[1:].lower()+first_letter.lower()+'ay')
```

```python
pig('dance');
```

```python
pig('Engineer')
```

7) File bloodtype1.txt records blood-types of patients (A, B, AB, O or OO) at a clinic. Write a function bldcount() that reads the file with name name and reports (i.e., prints) how many patients there are in each bloodtype.

```python
def bldcount(filename):
    
    #file read 
    x = open(filename,'r')
    words = x.readline()
    #creating empty list to store array
    abc=[]
    bloodtype_array = ['A','B','AB','O','OO']
   
    #Spliting the words and storing them in a list
    abc.append(words.split(" "))
    
    #checking number of patients for each blood type
    for blood_type in bloodtype_array:
        print("Blood type {} has {} patients".format(blood_type,abc[0].count(blood_type)))
```

```python
bldcount('bloodtype.txt')
```

8) Write a function curconv() that takes as input:a currency represented using a string (e.g., 'JPY' for the Japanese Yen or 'EUR' for the Euro)an amount and then converts and returns the amount in US dollars.

```python
def curconv(currency,amount):
    currencies = {}
    

    currency_file = open('currencies.txt')
    lines = currency_file.readlines()
    

    for x in lines:
        currencies[x[:3].strip()] = x[4:].split("\t")
    value = currencies[currency][0].strip()
    print(amount * float(value))
```

```python
curconv('AUD', 100)
```

```python
curconv('PHP', 100)
```

10) Encryption is the process of hiding the meaning of a text by substituting letters in the message with other letters, according to some system. If the process is successful, no one but the intended recipient can understand the encrypted message. Cryptanalysis refers to attempts to undo the encryption, even if some details of the encryption are unknown (for example, if an encrypted message has been intercepted). The first step of cryptanalysis is often to build up a table of letter frequencies in the encrypted text. Assume that the string letters is already defined as 'abcdefghijklmnopqrstuvwxyz'. Write a function called frequencies() that takes a string as its only parameter, and returns a list of integers, showing the number of times each character appears in the text. Your function may ignore any characters that are not in letters.

```python
def frequencies(text):
    
  
    alphabets = ('a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z')
    total_characters = []
    alpha_dict = {}
   
    for word in text:
        for char in word:
            if char.isalpha():
                total_characters.append(char)
   
    for char in alphabets:
        if char in total_characters:
            alpha_dict[char] = total_characters.count(char)
        else:
            alpha_dict[char] = 0
            
    #number of the alphabets present in the total_character list
    print(list(alpha_dict.values()))
```

```python
frequencies('Old Mcdonalds had a farm')
```

```python
frequencies('Eraser')
```
